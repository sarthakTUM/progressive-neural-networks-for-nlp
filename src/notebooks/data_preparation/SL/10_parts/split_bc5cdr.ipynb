```python
import pandas as pd
import numpy as np
import os
```

```python
random_state = 42
```

```python
def dataframe(X_file, y_file, id_file):

    X_list = []
    y_list = []
    id_list = []
    with open(X_file, 'r') as X, open(y_file, 'r') as y, open(id_file, 'r') as id:
        X_lines = X.readlines()
        y_lines = y.readlines()
        id_lines = id.readlines()
        assert len(X_lines) == len(y_lines)
        assert len(X_lines) == len(id_lines)
        for rec_X, rec_y, rec_id in zip(X_lines, y_lines, id_lines):
            id_list.append(rec_id.strip())
            X_list.append(rec_X.strip())
            y_list.append(rec_y.strip())
    return pd.DataFrame({
        'sentence':X_list,
        'label':y_list,
        'id':id_list
    })
        
```

```python
def write_df_SL(df, out):
    with open(os.path.join(out, 'sentences.txt'), 'w') as s, open(os.path.join(out, 'labels.txt'), 'w') as l, open(os.path.join(out, 'ID.txt'), 'w') as i:
        for row in df.itertuples():
            s.write(row.sentence+'\n')
            l.write(row.label+'\n')
            i.write(row.id+'\n')
```

```python
train_df = dataframe(X_file='../../../../../src/ner/data/bc5cdr_iobes_id/train/sentences.txt',
                    y_file='../../../../../src/ner/data/bc5cdr_iobes_id/train/labels.txt',
                    id_file = '../../../../../src/ner/data/bc5cdr_iobes_id/train/ID.txt')
```

```python
fracs = np.arange(0.1, 1.1, 0.1)
fracs
```

```python
root = '../../../../../src/ner/data/bc5cdr_iobes_id/train'
```

```python
for f in fracs:
    print('current fraction: {}'.format(str(f)))
    sample_df = train_df.sample(frac = f)
    print('data shape: {}'.format(sample_df.shape))
    out_file = os.path.join(root, 'frac_'+str(f))
    print('out_file: {}'.format(out_file))
    os.mkdir(out_file)
    write_df_SL(sample_df, out_file)
```
